```python
import pandas as pd
import numpy as np
import seaborn as sns
from bs4 import BeautifulSoup
from Bio import Entrez
```

```python
def search_pubmed_by_code(code):
    handle = Entrez.esearch(db="pubmed", term=code)
    record = Entrez.read(handle)
    handle.close()
    return record
```

```python
file_path = 'nctids.txt'

# Open the file and read the codes line by line
with open(file_path, 'r') as file:
    for line in file:
        # Remove any leading or trailing whitespace
        code = line.strip()
        
        # Search PubMed using the code
        try:
            record = search_pubmed_by_code(code)
            print(f"Code: {code}, PubMed search result: {record['Count']} articles found")
        except Exception as e:
            print(f"Error searching PubMed for code {code}: {e}")
```
