# Publication Linking

This notebook uses NCT identifiers from the downloaded summary table to look for linked PubMed records.

```python
from pathlib import Path
import pandas as pd

from trial_retention_toolkit import search_pubmed_for_codes

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
summary_path = ROOT / 'data/processed/mental_health_study_summary.csv'
df = pd.read_csv(summary_path)
sample_codes = df['nctId'].dropna().head(25).tolist()
sample_codes[:5]
```

```python
pubmed_df = pd.DataFrame(search_pubmed_for_codes(sample_codes))
pubmed_df.head()
```
