# Mental Health Study Landscape

This notebook explores the downloaded ClinicalTrials.gov summary table.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
summary_path = ROOT / 'data/processed/mental_health_study_summary.csv'
df = pd.read_csv(summary_path)
df.head()
```

```python
df[['overallStatus', 'studyType', 'hasResults']].describe(include='all')
```

```python
status_counts = df['overallStatus'].fillna('Unknown').value_counts().head(15)

plt.figure(figsize=(10, 6))
status_counts.sort_values().plot(kind='barh')
plt.title('Most Common Overall Status Values')
plt.xlabel('Study count')
plt.ylabel('Status')
plt.tight_layout()
```

```python
results_rate = (
    df.assign(hasResults=df['hasResults'].fillna(False))
      .groupby('studyType')['hasResults']
      .mean()
      .sort_values(ascending=False)
)

results_rate
```

```python
country_counts = (
    df['countries']
      .dropna()
      .str.split('; ')
      .explode()
      .value_counts()
      .head(15)
)

country_counts
```
