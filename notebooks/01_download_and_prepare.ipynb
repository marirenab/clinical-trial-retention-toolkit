# Download And Prepare ClinicalTrials.gov Data

This notebook downloads mental-health studies from the ClinicalTrials.gov v2 API and creates a tidy summary table for downstream analysis.

```python
from pathlib import Path
import json
import pandas as pd

from trial_retention_toolkit import build_study_summary_frame, fetch_studies, save_study_download
```

```python
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
raw_path = ROOT / 'data/raw/mental_health_studies.json'
summary_path = ROOT / 'data/processed/mental_health_study_summary.csv'

payload = fetch_studies(query_cond='mental health', page_size=1000)
save_study_download(payload, raw_path)

summary_df = build_study_summary_frame(payload['studies'])
summary_df.to_csv(summary_path, index=False)

payload['downloadedCount'], summary_df.shape
```
